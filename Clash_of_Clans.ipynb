```python
# api request
request = requests.get("https://set7z18fgf.execute-api.us-east-1.amazonaws.com/prod/?route=getClanDetails&clanTag=%23VQPVPJ09")
```

```python
# request in eine text datei umwandeln
request_text = request.text
```

```python
# json daten laden als dictionary
data=json.loads(request_text)
```
